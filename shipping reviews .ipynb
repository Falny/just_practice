```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://krasnoyarsk.allcafe.ru/rating/top50/')
```

```python
response.status_code
```

```python
response.headers
```

```python
content = response.content
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
top_rest = soup.find_all('div', attrs={'class' : 'placeList placeList-default'})
list_tr = top_rest[0].find_all('div', attrs={'class' : 'placeList_item'})
```

```python
list_rest = []
for tr in list_tr:
    dataframe={}
    dataframe['rest_name'] = (tr.find('div', attrs={'class' : 'placeList_title'})).text.replace('\n', ' ')
    dataframe['resr_adress'] = (tr.find('span', attrs={'class' : 'placeList_addr'})).text.replace('\n', ' ')
    dataframe['cuisine_type'] = (tr.find('span', attrs={'class' : 'placeList_info'})).text.replace('\n', ' ')
    list_rest.append(dataframe)
list_rest
```

```python
df = pd.DataFrame(list_rest)
df.to_csv('kras_restoran', index = False)
```
