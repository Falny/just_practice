```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://www.sravni.ru/kursy/analitika-data-science/')
response.status_code
```

```python
response.headers
```

```python
content = response.content
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
main = soup.find_all('div', attrs = {'class' : 'style_wrapper__OiJ9S style_wrapperCell__vn1mh'})
list_one = main[0].find_all('div', attrs = {'class' : 'style_wrapper__JdiJD style_card__ovlfm'})
```

```python
list_all = []

for l in list_one:
    dataframe = {}
    dataframe['name'] = (l.find('div', attrs = {'class':'style_heading__WaLQK style_headingTopPadding__sB1tl'})).text.replace('\n', ' ')
    dataframe['installments'] = (l.find('div', attrs = {'class':'style_value__zeZ0y style_value1__0R9hm'})).text.replace('\n', ' ')
    dataframe['price'] = (l.find('div', attrs = {'class':'style_value__zeZ0y style_value2__0I_72'})).text.replace('\n', ' ')
    list_all.append(dataframe)
list_all
```

```python
df = pd.DataFrame(list_all)
df.to_csv('course', index = False)
```
